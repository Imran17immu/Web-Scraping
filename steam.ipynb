```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
base_site = 'https://store.steampowered.com/search/?sort_by=Released_DESC&tags=19'
response = requests.get(base_site)
html = response.content
```

```python
response
```

```python
soup = BeautifulSoup(html , 'lxml')
```

```python
soup
```

```python
print(type(soup))
```

```python
with open('steam.html', 'wb')as file:
    file.write(soup.prettify('utf-8'))
```

```python
divs = soup.find_all('div', {'id':'search_resultsRows'})
```

```python
divs
```

```python
titles =[]
for tt in divs:
    tt.find_all('span', {'class':'title'})
    titles.append(tt.find_all('span', {'class':'title'}))

titles
```

```python
title = []
for tt in titles:
    for t in tt:
        t.string
        title.append(t.string)
title        
```

```python
len(title)
```

```python
#year
```

```python
years = []
for y in divs:
    y.find_all('div',{'class':'col search_released responsive_secondrow'})
    years.append(y.find_all('div',{'class':'col search_released responsive_secondrow'}))
    
print(years)    
```

```python
len(years[0])
```

```python
year = []
for yy in years:
    for y in yy:
        y.string
        year.append(y.string)
        
        
year        
```

```python
len(year)
```

```python
#discount
```

```python
discounts = []
for dd in divs:
    dd.find_all('div', {'class':'col search_discount responsive_secondrow'})
    discounts.append(dd.find_all('div', {'class':'col search_discount responsive_secondrow'})) 
    
print(discounts)    
```

```python
discount = []
for dd in discounts:
    for d in dd:
        if d.find('span') == None:
            discount.append(None)
        else:
            discount.append(d.find('span').string)
            
discount   
        
       
```
